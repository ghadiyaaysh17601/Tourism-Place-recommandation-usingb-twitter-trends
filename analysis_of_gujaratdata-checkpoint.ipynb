```python
import numpy as np # numerical computing 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #visualization
import seaborn as sns #modern visualization
```

```python
#load csv file
df=pd.read_csv(r"C:\Users\Ayush Patel\Downloads\data2.csv")
```

```python
plt.figure(figsize=(15,5))
sns.countplot('Location', data = df)
plt.title("Number of Tweet on specific location",fontsize=18,fontweight="bold")
plt.ylabel("Tweet", size = 25)
plt.xlabel("Places", size = 25)
plt.xticks(size = 20)
plt.yticks(size = 20)
```

```python
MoM= df['Location'].value_counts()
MoM.head(10).plot(kind = 'bar',figsize=(12,8), fontsize=15, color='blue')
plt.title("Top 10 cities to visit in Gujarat",fontsize=18,fontweight="bold")
plt.ylabel("Frequency", size = 25)
plt.xlabel("Cities", size = 25)
```

```python
df['Location'].value_counts().plot(kind='pie', fontsize=14, autopct='%3.1f%%', 
                                               figsize=(10,7), shadow=True, startangle=135, legend=True, cmap='Oranges')


plt.title('Percentage of tweet on specific location')
```
